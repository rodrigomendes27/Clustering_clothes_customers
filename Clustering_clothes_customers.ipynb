# Clusterização dos clientes na tabela estilos

### Função para fazer o primeiro K-means

Função que recebe um conjunto de dados, calcula o KMeans para 19 quantidades de clusters que vão de 2 a 20 possíveis 
agrupamentos e, finalmente, retorna uma lista com o nosso wcss

```python
from sklearn.cluster import KMeans
def calculate_wcss(data):

    wcss = []
    for n in range(2, 10):
        kmeans = KMeans(n_clusters=n)
        kmeans.fit(X=data)
        wcss.append(kmeans.inertia_)

    return wcss
```

Esse método optimal_number_of_clusters() recebe uma lista contendo as somas dos quadrados para cada quantidade de clusters que calculamos usando calculate_wcss() e, como resultado, retorna a quantidade ótima de clusters.

```python
import numpy as np

def optimal_number_of_clusters(wcss):
    x1, y1 = 2, wcss[0]
    x2, y2 = 20, wcss[len(wcss)-1]
    distances = []
    for i in range(len(wcss)):
        x0 = i+2
        y0 = wcss[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = np.sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distances.append(numerator/denominator)
    
    return distances.index(max(distances)) + 2
```

### Jeito Novo Unificado

```python
from sklearn.cluster import KMeans

# calculando a soma dos quadrados para as 19 quantidade de clusters
wcss_result = calculate_wcss(cod_df_quest_fem_2)

# calculando a quantidade ótima de clusters
n = optimal_number_of_clusters(wcss_result)

# rodando o kmeans para nossa quantidade ótima de clusters
kmeans_clientes = KMeans(n_clusters=n)
clusters = kmeans_clientes.fit_predict(cod_df_quest_fem_2)

# adicionando os resultado na tabela
cod_df_quest_fem_2["CLUSTER CLIENTES"] = clusters
```
